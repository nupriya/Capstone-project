```python
driver.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc")
```

# WebScrapping 

```python
Sno=[]
movie_name=[]
duration=[]
year=[]
ratings=[]
metascore=[]
votes=[]
genre=[]
gross_collection=[]
certification=[]
director_name=[]
stars_name=[]
```

```python
start=0
end=35
delay=10
for page in range(start,end):
    WebDriverWait(driver,delay)
        
    try:
        no=driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
        for i in no:
            Sno.append(i.text)
    except NoSuchElementException:
        Sno.append('-')
    except StaleElementReferenceException:
        Sno.append('-')
        
    try:
        mname=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]')
        for j in mname:
            movie_name.append(j.text.split()[1:-1])
    except NoSuchElementException:
        movie_name.append('-')
    except StaleElementReferenceException:
        movie_name.append('-')   
        
    try:
        y=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
        for j in y:
            year.append(j.text)
    except NoSuchElementException:
        year.append('-')
    except StaleElementReferenceException:
        year.append('-')
        
    try:
        dura=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
        for j in dura:
            duration.append(j.text)
    except NoSuchElementException:
        duration.append('-')
    except StaleElementReferenceException:
        duration.append('-')
        
    try:
        rate=driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]')
        for j in rate:
            ratings.append(j.text)
    except NoSuchElementException:
        ratings.append('-')
    except StaleElementReferenceException:
        ratings.append('-')     
    
    
    try:
        meta=driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-metascore"]')
        for j in meta:
            metascore.append(j.text.strip())
    except NoSuchElementException:
        metascore.append('None')
    except StaleElementReferenceException:
        metascore.append('-') 
      
        
    try:
        vote=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
        for j in vote:
            votes.append(j.text.strip().split()[1])
    except NoSuchElementException:
        votes.append('-')
    except StaleElementReferenceException:
        votes.append('-')
        
        
    try:
        gross=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
        for j in gross:
            gross_collection.append(j.text.split()[-1])
    except NoSuchElementException:
        gross_collection.append('-')
    except StaleElementReferenceException:
        gross_collection.append('-')  
        
        
    try:
        gen=driver.find_elements(By.XPATH,'//span[@class="genre"]')
        for j in gen:
            genre.append(j.text)
    except NoSuchElementException:
        genre.append('-')
    except StaleElementReferenceException:
        genre.append('-')
        
    try:
        certi=driver.find_elements(By.XPATH,'//span[@class="certificate"]')
        for j in certi:
            certification.append(j.text)
    except NoSuchElementException:
        certification.append('Not Available')
    except StaleElementReferenceException:
        certification.append('-')  
        
    try:
        link=driver.find_elements(By.XPATH,'//*[@id="main"]/div/div[3]/div/div[position()>=1 and position()<=1750]/div[3]/p[3]')
        for i in link:
            director_name.append(i.text.strip()[10:].split('|')[0:1])
            stars_name.append(i.text.strip()[15:].split('|')[1:])
    except NoSuchElementException:
        director_name.append('-')
        stars_name.append('-')
    except StaleElementReferenceException:
        director_name.append('-') 
        stars_name.append('-')
        
        
    try:
        next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
        next_button.click()
    except NoSuchElementException:
        pass
    
        
            
```

```python
driver.close()
```
